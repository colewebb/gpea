# Manual image analysis

```python
import os
import numpy as np
from matplotlib.pyplot import imshow
from PIL import Image
%matplotlib inline

os.chdir("C:/Users/Admin/Documents/Pi2-selected")
threshold = 140
currentWhite = 0
oldWhite = 0
currentNumpy = np.array(Image.open("./2021817130.png"))
oldNumpy = np.array(Image.open("./2021816130.png"))
imshow(currentNumpy)
```

```python
imshow(oldNumpy)
```

```python
def analyze(array):
    interestingPixels = 0
    for i in array:
        for j in i:
            if (int(j[0]) + int(j[1]) + int(j[2]))/3 > threshold:
                j[0] = 255
                j[1] = 255
                j[2] = 255
                interestingPixels += 1
            else:
                j[0] = 0
                j[1] = 0
                j[2] = 0
    return interestingPixels

currentWhite = analyze(currentNumpy)
print(currentWhite)
imshow(currentNumpy)
```

```python
oldWhite = analyze(oldNumpy)
print(oldWhite)
imshow(oldNumpy)
```
