# Manual image analysis

```python
import os
import numpy as np
import cv2 as cv
from matplotlib.pyplot import imshow
from PIL import Image
%matplotlib inline

os.chdir("C:/Users/Admin/Downloads")
threshold = 150
currentBlack = 0
oldBlack = 0
current = cv.imread("./10-4-2021 17 0.png")
old = cv.imread("./10-5-2021 17 0.png")
imshow(current)
```

```python
imshow(old)
```

```python
def analyze(picture):
    arrayCV = cv.cvtColor(picture, cv.COLOR_BGR2GRAY)
    thresholdedValue, arrayCV = cv.threshold(arrayCV, 0, 1, cv.THRESH_OTSU)
    return [thresholdedValue, arrayCV]

currentThreshold, currentThresholded = analyze(current)
imshow(currentThresholded)
```

```python
oldThreshold, oldThresholded = analyze(old)
imshow(oldThresholded)
```

```python
for i in range(0, len(current)):
    for j in range(0, len(current[0])):
        if currentThresholded[i, j] == 1:
            currentBlack += 1
for i in range(0, len(current)):
    for j in range(0, len(current[0])):
        if oldThresholded[i, j] == 1:
            oldBlack += 1
print(currentBlack, oldBlack)
```
