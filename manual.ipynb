# Manual image analysis

```python
import os
import numpy as np
from matplotlib.pyplot import imshow
from PIL import Image
%matplotlib inline

os.chdir("C:/Users/Admin/Downloads")
threshold = 180
currentWhite = 0
oldWhite = 0
currentNumpy = np.array(Image.open("./2021823160.png").crop((1000, 300, 2300, 2200)))
oldNumpy = np.array(Image.open("./2021822160.png").crop((1000, 300, 2300, 2200)))
imshow(currentNumpy)
```

```python
imshow(oldNumpy)
```

```python
def analyze(array):
    interestingPixels = 0
    for i in array:
        for j in i:
            if (int(j[0]) + int(j[1]) + int(j[2]))/3 > threshold:
                j[0] = 255
                j[1] = 255
                j[2] = 255
                interestingPixels += 1
            else:
                j[0] = 0
                j[1] = 0
                j[2] = 0
    return interestingPixels

currentWhite = analyze(currentNumpy)
print(currentWhite)
imshow(currentNumpy)
```

```python
oldWhite = analyze(oldNumpy)
print(oldWhite)
imshow(oldNumpy)
```

```python
print(str(currentWhite - oldWhite))
print(str(((currentWhite/oldWhite) - 1 )* 100)[:6] + "%")
```
