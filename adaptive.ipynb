# Adaptive Thresholding Testing

```python
import cv2 as cv
import os
from matplotlib import pyplot
%matplotlib inline
img = cv.cvtColor(cv.imread("C:/Users/Admin/Downloads/img.png"), cv.COLOR_BGR2GRAY)[500:2000, 500:2400]
ret, thresholded = cv.threshold(img, 0, 1, cv.THRESH_OTSU)
pyplot.imshow(thresholded)
```

```python
interesting = 0
for i in range(len(thresholded)):
    for j in range(len(thresholded[0])):
        if thresholded[i, j] == 1:
            interesting += 1
interesting
```
